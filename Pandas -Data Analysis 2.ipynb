## CHALLENGE -10 - In every olympics, which US athlete has won the most total number of medals? Include the atheletes discipline


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
oo=pd.read_csv(r"C:\Users\n.rahman\OneDrive - BALADNA\Desktop\olympics.csv")
oo.head()
```

```python
gy = oo[oo.NOC=='USA']
df = gy.groupby(['Edition','Athlete','Discipline','Medal']).size().unstack('Medal',fill_value=0)
df
#create a column or series called 'total' to get the total sum of medals
df['Total'] = df['Gold'] + df['Silver'] + df['Bronze']
df
```

```python
#now we need to determine for each edition who is US athlete who won the most no of medals
#for that we need to reset the index first
df.reset_index(inplace=True)
df
```

```python
#now we work our way to each of the groups created by the groupby object. This splits for every edition
for year,group in df.groupby('Edition'):
    print(group.sort_values('Total',ascending=False)[:1]) #now sort the values in ascending and slice the top most one which is required
```

```python
#Now the easiest way to store this all in a list. Reason is then we can populate the dataframe from element in the list
top =[group.sort_values('Total',ascending=False)[:1] for year,group in df.groupby('Edition')]
top
```

```python
#create a new dataframe called topusa
topusa = pd.DataFrame()
for i in top:
    topusa = topusa.append(i)

topusa
```
